# Wykres czasu działania programu w zależności od wielkości Bufora.

## Rozwiązanie bez zachowania kolejności.

### Producent

```python
plt.figure(figsize=(20,10))
x = [10, 20, 50 ,100, 200] # średnia z 10 pomiarów
y = [1675275, 1666080, 1679213, 1671344, 1682355] # mikrosekujndy
plt.scatter(x,y,label='wykres', marker='*', s = 200)
plt.xlabel('Wielkość bufora') 
# Labeling the Y-axis 
plt.ylabel('Czas działania (mikrosekundy)') 
# Give a title to the graph
plt.title('Wykres czasu działania programu w zależności od wielkości bufora') 
 
plt.show() 
```

# Porównanie czasu działania dla różnych wersji.

## Producent

```python
plt.figure(figsize=(400000, 100))
y = [[1675275, 1666080, 1679213, 1671344, 1682355],
[1511178, 2573141, 6413765, 12049032, 24333555],
 [181196, 93926, 41862, 33848, 32447]]

X = np.arange(5)
fig, ax = plt.subplots(figsize=(18,10))
ax.bar(X + 0.00, y[0], color = 'b', width = 0.25, label="UNORDERED")
ax.bar(X + 0.25, y[1], color = 'g', width = 0.25, label="ORDERED")
ax.bar(X + 0.50, y[2], color = 'r', width = 0.25, label="NO-SCP")
ax.set_xticklabels(['','Rozmiar bufora = 10','Rozmiar bufora = 20','Rozmiar bufora = 50','Rozmiar bufora = 100','Rozmiar bufora = 200'])

plt.legend()
plt.show() 
```

## Konsument

```python
plt.figure(figsize=(400000, 100))
y = [[1671032, 1642123, 1680012, 1677723, 1681323],
[1523178, 2358213, 6892132, 12234321, 22431234],
 [189391, 95123, 40312, 35312, 30432]]

X = np.arange(5)
fig, ax = plt.subplots(figsize=(18,10))
ax.bar(X + 0.00, y[0], color = 'b', width = 0.25, label="UNORDERED")
ax.bar(X + 0.25, y[1], color = 'g', width = 0.25, label="ORDERED")
ax.bar(X + 0.50, y[2], color = 'r', width = 0.25, label="NO-SCP")
ax.set_xticklabels(['','Rozmiar bufora = 10','Rozmiar bufora = 20','Rozmiar bufora = 50','Rozmiar bufora = 100','Rozmiar bufora = 200'])
 
plt.legend()
plt.show() 
```
