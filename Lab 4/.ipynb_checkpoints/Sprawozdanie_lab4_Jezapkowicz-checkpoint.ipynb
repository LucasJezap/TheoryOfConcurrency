```python
data = pd.read_csv("results.csv")
data
```

Teraz postaramy się przedstawić wszystkie konfiguracje na wykresie typu box-plot. Do przedstawienia mamy 16 konfiguracji (razy 2 bo producenci i konsumenci).  
Warto zacząć od przygotowania danych oraz funkcji tworzącej wykresy.

```python
tmp_data = data.groupby(["Bufsize", "PC_config", "Is_fair", "Randomization", "Prod_or_cons"])
data_to_draw = []
```

```python
for info, group_data in tmp_data:
    data_to_draw.append((info, group_data))
```

```python
plt.rcParams["figure.figsize"] = (15,5)

def draw_plot(config, data):
    x = []
    y = []
    for i,j in data.items():
        x.append(str(i))
        y.append(j)
    plt.scatter(x,y)
#     plt.figure(figsize=(10,5))
    plt.show()

def draw(draw_data, index, multiplier=1, print_more=False):
    print("Configuration number " + str(index+1))
    print("Buffer size: " + str(draw_data[0][0]))
    print("Type: " + str(draw_data[0][4]))
    print("PC Config: " + str(draw_data[0][1]))
    print("Is buffer fair: " + str(draw_data[0][2]))
    print("Distribution: " + str(draw_data[0][3]))
    print("Number of buckets: " + str(50 / multiplier))
    
    min_size = 0
    max_size = 5000
    interval_size = 100
    if draw_data[0][0] > 10000:
        interval_size = 1000
        max_size = 50000
    
    if interval_size == 100:
        draw_data[1]['bin'] = pd.cut(draw_data[1]['Size'], [x for x in range(min_size, max_size + 99, interval_size * multiplier)])
    else:
        draw_data[1]['bin'] = pd.cut(draw_data[1]['Size'], [x for x in range(min_size, max_size + 999, interval_size * multiplier)])
    a = draw_data[1].groupby('bin')["Accesses"].sum()
    a.plot.box()
    plt.show()
    if print_more:
        print()
        draw_plot(draw_data[0] ,a)
```

Wszystkie eksperymenty przeprowadzone są dla 50 kubełków.  
Postaramy się pokazać na wykresach różnice między konfiguracjami.

## 1. Producenci

Zaczniemy od narysowania wszystkich wykresów pudełkowych dla producentów (16 wykresów).

Na pierwszy rzut oka widać, że dla dystrybucji uniform pudełka są zazwyczaj większe i nie mają (lub mają mało wartości odstających). Przyjrzyjmy się teraz bliżej poszczególnym parametrom.

### Wielkość bufora

```python
draw(data_to_draw[1], 1)
draw(data_to_draw[17], 17)
print('--------------------------------------------------------------------------------------------------------------------')
print('--------------------------------------------------------------------------------------------------------------------')
print('--------------------------------------------------------------------------------------------------------------------')
draw(data_to_draw[5], 5)
draw(data_to_draw[21], 21)
```

Wniosek: Czym większa większość bufora tym większa średnia ilość dostępów do bufora. Kształt wykresu jest jednak bardzo podobny - ilość buforów rozkłada się zatem podobnie (oczywiście z przeskalowaniem).
