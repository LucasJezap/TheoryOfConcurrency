```python
import matplotlib.pyplot as plt
import numpy as np
```

# Wykres czasu działania programu w zależności od liczby wątków.

```python
plt.figure(figsize=(20,10))
x = [1,3,5,10,20,50,100,200,500,1000]
y = [4463,1388,1187,1095,1061,1038,1068,1126,44,95]
plt.plot(x,y,label='wykres')
plt.xlabel('Ilość wątków') 
# Labeling the Y-axis 
plt.ylabel('Czas działania') 
# Give a title to the graph
plt.title('Wykres czasu działania programu w zależności od liczby wątków') 
  
# Show a legend on the plot 
plt.legend() 
 
plt.show() 
```
